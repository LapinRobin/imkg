```python
import json, os
from os.path import exists
```

```python
def save2json(filename, dump):
    out_file = open(filename, "w+")
    json.dump(dump, out_file)
    out_file.close()
```

```python
f1 = open('../../imgflip.cleaned.json', 'r', encoding='utf8')
j1 = json.load(f1)
```

```python
j1
```

```python
existing = set([file.replace('.json','') for file in os.listdir('./') if (os.path.isfile(os.path.join('./', file)) and file.endswith("json"))])
```

```python
all_templates = set([j1[k]['template_ID'] for k in j1.keys()])
```

```python
instances  = [j1[k] for k in j1.keys() ]
```

```python
j = 0
for t in list(all_templates - existing):
    f = open(t+".json", "w")
    json.dump([i for i in instances if i['template_ID']==t], f)
    j = j + 1
    print(str(len(os.listdir('./'))) + ' elements over ' + str(len(all_templates)) + ' processed' )
```

```python
templates = set()
for k in j1.keys():
    meme = j1[k]
    if(exists(meme['template_ID']+".json")):
        f = open(meme['template_ID']+".json", "r")
        js = json.load(f)
        if(meme not in js):
            js.append(meme)
            json.dump(js, open(meme['template_ID']+".json", "w"))
    templates.add(meme['template_ID'])
    print(str(len(templates)) + ' elements over ' + str(all_templates) + ' processed' )
```
