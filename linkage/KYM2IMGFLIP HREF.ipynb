```python
import json, csv
from csv import writer, reader
```

```python
def save2json(filename, dump):
    out_file = open(filename, "w")
    json.dump(dump, out_file, indent = 6)
    out_file.close()
```

```python
def save2csv(filename, header, data):
    with open(filename, 'w', newline='') as csvfile:
        csvw = csv.writer(csvfile, delimiter=';',quotechar='|', quoting=csv.QUOTE_MINIMAL)
        csvw.writerow(header)
        for d in data:
            csvw.writerow(d)
```

```python
kymj = open('../kym/kym.memes.json')
data = json.load(kymj)
```

```python
links = []
for d in data:
    if 'content' in d.keys():
        c = d['content']
        for k in c.keys():
            if 'links' in c[k].keys():
                for l in c[k]['links']:
                        if ';' not in l[1]:
                            links.append([d['url'],l[1]])

save2csv("kym_hrefs.csv",['meme_url',"link"],links)
```
