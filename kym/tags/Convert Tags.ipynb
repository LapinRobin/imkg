```python
import json, csv, os
```

```python
def save2json(filename, dump):
    out_file = open(filename, "w")
    json.dump(dump, out_file, indent = 6)
    out_file.close()
```

```python
tagsfile = open('./kym.memes.tags.spotlight.json')
```

```python
jsonfile = open('../dbpedia-wikidata.json')
```

```python
jqid = json.load(jsonfile)
```

```python
jtags = json.load(tagsfile)
```

```python
res = list()
```

```python
mfile = open('./memes.csv')
memes = [m[0] for m in csv.reader(mfile)]    
```

```python
memes = memes[1:len(memes)]
```

```python
memes 
```

```python
for url in jtags.keys():
    if 'Resources' in jtags[url].keys() and url in memes:
        entities = list(set([e["@URI"].replace('http://dbpedia.org/resource/','') for e in jtags[url]['Resources']]))
        wiki_entities = ['https://www.wikidata.org/wiki/'+jqid[e] for e in entities if e in jqid.keys()]
        res.append({'url': url, 'tags_entities': wiki_entities})
```

```python
save2json('./tags_entities_selected.json',res)
```
